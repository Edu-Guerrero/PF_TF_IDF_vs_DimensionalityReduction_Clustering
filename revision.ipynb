# TF_IDF #

```python
import pandas as pd
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('tf_idf_documentos.csv')

# Sort the dataframe by the first column
df = df.sort_values(by=df.columns[0])

# Omit the first column which contains the names of the files
tf_idf_data = df.iloc[:, 1:]

# Compute cosine similarity
cosine_sim = cosine_similarity(tf_idf_data)

# Create a heatmap for cosine similarity
plt.figure(figsize=(10, 8))
sns.heatmap(cosine_sim, cmap='viridis', annot=False)
plt.title('Heatmap of Cosine Similarity of TF-IDF Vectors')
plt.xlabel('Documents')
plt.ylabel('Documents')
plt.show()
```

```python
from sklearn.metrics.pairwise import euclidean_distances

# Compute Euclidean distance
euclidean_dist = euclidean_distances(tf_idf_data)

# Create a heatmap for Euclidean distance
plt.figure(figsize=(10, 8))
sns.heatmap(euclidean_dist, cmap='viridis', annot=False)
plt.title('Heatmap of Euclidean Distance of TF-IDF Vectors')
plt.xlabel('Documents')
plt.ylabel('Documents')
plt.show()
```

```python
from sklearn.metrics.pairwise import manhattan_distances

# Compute Manhattan distance
manhattan_dist = manhattan_distances(tf_idf_data)

# Create a heatmap for Manhattan distance
plt.figure(figsize=(10, 8))
sns.heatmap(manhattan_dist, cmap='viridis', annot=False)
plt.title('Heatmap of Manhattan Distance of TF-IDF Vectors')
plt.xlabel('Documents')
plt.ylabel('Documents')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
from sklearn.metrics import jaccard_score
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('tf_idf_documentos.csv')

# Omit the first column which contains the names of the files
tf_idf_data = df.iloc[:, 1:]

# Binarize the TF-IDF data
binary_tf_idf_data = tf_idf_data.applymap(lambda x: 1 if x > 0 else 0)

# Compute Jaccard similarity
jaccard_sim = pd.DataFrame(index=df.index, columns=df.index)
for i in range(len(binary_tf_idf_data)):
    for j in range(len(binary_tf_idf_data)):
        jaccard_sim.iloc[i, j] = jaccard_score(binary_tf_idf_data.iloc[i], binary_tf_idf_data.iloc[j])

# Create a heatmap for Jaccard similarity
plt.figure(figsize=(10, 8))
sns.heatmap(jaccard_sim.astype(float), cmap='viridis', annot=False)
plt.title('Heatmap of Jaccard Similarity of TF-IDF Vectors')
plt.xlabel('Documents')
plt.ylabel('Documents')
plt.show()
```

```python
# Compute Pearson correlation
pearson_corr = tf_idf_data.T.corr(method='pearson')

# Create a heatmap for Pearson correlation
plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, cmap='viridis', annot=False)
plt.title('Heatmap of Pearson Correlation of TF-IDF Vectors')
plt.xlabel('Documents')
plt.ylabel('Documents')
plt.show()
```
