# TF_IDF #

```python
import pandas as pd
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('tf_idf_documentos.csv')

# Sort the dataframe by the first column
df = df.sort_values(by=df.columns[0])

# Omit the first column which contains the names of the files
tf_idf_data = df.iloc[:, 1:]

# Compute cosine similarity
cosine_sim = cosine_similarity(tf_idf_data)

# Create a heatmap for cosine similarity
plt.figure(figsize=(10, 8))
sns.heatmap(cosine_sim, cmap='viridis', annot=False)
plt.title('Heatmap of Cosine Similarity of TF-IDF Vectors')
plt.xlabel('Documents')
plt.ylabel('Documents')
plt.show()
```

```python
from sklearn.metrics.pairwise import manhattan_distances

# Compute Manhattan distance
manhattan_dist = manhattan_distances(tf_idf_data)

# Create a heatmap for Manhattan distance
plt.figure(figsize=(10, 8))
sns.heatmap(manhattan_dist, cmap='viridis', annot=False)
plt.title('Heatmap of Manhattan Distance of TF-IDF Vectors')
plt.xlabel('Documents')
plt.ylabel('Documents')
plt.show()
```
